```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
# separete train data to X, y
X_train = df_train[["1stFlrSF", "TotalBsmtSF", "GarageArea", "GarageCars", "GrLivArea", "OverallQual"]]
y_train = df_train["SalePrice"]
```

```python
X_train.info()
```

```python
from sklearn.model_selection import train_test_split

X_train_pra, X_train_val, y_train_pra, y_train_val = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train_pra, y_train_pra)
```

```python
model.predict(X_train_val)
```

```python
model.score(X_train_pra, y_train_pra)
```

```python
model.score(X_train_val, y_train_val)
```
