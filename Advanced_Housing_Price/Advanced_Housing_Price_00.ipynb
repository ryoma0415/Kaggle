```python
df_X_train_obj = pd.DataFrame()
df_X_train_int = pd.DataFrame()
df_X_train_float = pd.DataFrame()

for col in X_train_.columns:
  if X_train_[col].dtype == "O":
    df_X_train_obj = pd.concat([df_X_train_obj, X_train_[col]], axis=1)
  elif X_train_[col].dtype == "int64":
    df_X_train_int = pd.concat([df_X_train_int, X_train_[col]], axis=1)
  else:
    df_X_train_float = pd.concat([df_X_train_float, X_train_[col]], axis=1)
```
